# MatplotLib 
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
#plotting using matplotlib 
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph: ')
```

```python
y=x*x
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,21) 
y = 3 * x ** 2 + 5 * x + 6
plt.title("Demo") 
plt.xlabel("x axis ") 
plt.ylabel("y axis ") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute the x and y coordinates on sine and cosine curves 
x = np.arange(0, 2 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
   
# Make the first plot 

plt.subplot(2, 1, 1)
plt.plot(x, y_sin) 
plt.title('Sine')  
   
# make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,) 
plt.title('Cosine')  



# Show the figure. 
plt.show()
```

# BARS

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "blue", width = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color = "blue", height = 0.5)
plt.show()
```

# HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x, bins = 50)
plt.show() 
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

plt.boxplot(data,vert=True,patch_artist=False) 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, explode= (0.1,0.1,0.1,0.1))
plt.legend()
plt.show()
```
