```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
df.corr()
```

```python
#Correlation with Heatmap
```

```python
sns.heatmap(df.corr())
```

# JoinPlot
A join plot allows to study the relationship between 2 numeric variables.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot
It is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

# DistPlot

```python
sns.distplot(df['tip'])
```

# CountPlot

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='smoker',hue="sex",data=df)
```

# BarPlot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# BoxPlot

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
sns.boxplot(x="tip", y="day", hue="smoker",data=df)
```

```python
sns.boxplot(x="total_bill", y="day", hue="sex",data=df)
```

```python
sns.boxplot(x="tip", y="day", hue="sex",data=df)
```

# Violin Plot
It help us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df)
```
